```python
import selenium
import pandas as pd
from selenium import webdriver
import os
```

```python
driver = webdriver.Chrome('chromedriver.exe')
```

```python
url = "https://flight.yatra.com/air-search-ui/dom2/trigger?ADT=1&CHD=0&INF=0&class=Economy&destination=BOM&destinationCountry=IN&flexi=0&flight_depart_date=20%2F11%2F2021&hb=&noOfSegments=1&origin=DEL&originCountry=IN&type=O&unique=1637333218941&viewName=normal"
driver.get(url)
```

```python
flight_name = driver.find_elements_by_xpath("//span[@class='i-b text ellipsis']")
flight_dep = driver.find_elements_by_xpath("//p[@class='fs-10 font-lightgrey no-wrap city ellipsis']")
flight_dep_time = driver.find_elements_by_xpath("//div[@class='i-b pr']")
flight_arr_time = driver.find_elements_by_xpath("//p[@class='bold fs-15 mb-2 pr time']")
flight_duration = driver.find_elements_by_xpath("//p[@class='fs-12 bold du mb-2']")
flight_journey_type = driver.find_elements_by_xpath("//span[@class='cursor-default']")
flight_Cost = driver.find_elements_by_xpath("//div[@class='i-b tipsy fare-summary-tooltip fs-18']")

Airline = []
Dep_arr_City = []
Departure = []
Arrival = []
dep_time = []
arr_time = []
duration = []
Journey_Type = []
Price = []


for i in flight_name:
    Airline.append(i.text)
for i in flight_dep:
    Dep_arr_City.append(i.text)
for i in range(0,len(Dep_arr_City),1):
    Departure.append(Dep_arr_City[0])
    Arrival.append(Dep_arr_City[1])
for i in flight_dep_time:
    dep_time.append(i.text)
for i in flight_arr_time:    
    arr_time.append(i.text)
for i in flight_duration:     
    duration.append(i.text)
for i in flight_journey_type:
    Journey_Type.append(i.text)
for i in flight_Cost:
     Price.append(i.text)   
        
df_flight = pd.DataFrame({'Airline':Airline[0:51] ,'Departure':Departure[0:51] ,'Arrival':Arrival[0:51],'Dep_time':dep_time[0:51],'Arr_time':arr_time[0:51],
                        'Duration':duration[0:51],'Journey_Type':Journey_Type[0:51],'Price':Price[0:51]})        
```
